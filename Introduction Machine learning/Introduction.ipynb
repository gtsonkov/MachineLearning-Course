```python
for group_id, data in titanic_data.groupby("Survived"):
    print(data.Fare)
    break
```

```python
for group_id, data in titanic_data.groupby("Survived"):
    fares = data.Fare
    fares = fares[fares>0]
    log_fares = np.log(fares)
    plt.hist(log_fares, alpha = 0.7, bins = "fd", density = 1, label = group_id)
    plt.xlabel("ln(Fare)")
    plt.ylabel("Frequency (Normalized)")

    plt.legend()
```
